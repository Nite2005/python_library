```python
import matplotlib.pyplot as plt
```

Linear Graph

```python
x=[1,2,3,4,5]
y=[5,6,7,8,9]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[5,6,7,8,9]
c='r'
plt.plot(x,y,c)
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y,'Black')
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y,marker = 'o',)
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.plot(y,marker='>')
plt.plot(x,marker='o')
plt.show()
```

Bar Graph

```python
x=[1,2,3,4,5]
y=[5,6,7,8,9]
plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[5,6,7,8,9]
c = ['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
import numpy as np
```

```python
# making graph in horizontal direction
x= np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
plt.barh(x,y,color='r')
plt.show()
```

```python
x= np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
plt.bar(x,y,color='g',width = 0.2)
plt.show()
```

```python
x= np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
plt.barh(x,y,color='g',height=0.1)
plt.show()
```

Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,6]
plt.title("Scatter plot")
plt.xlabel("month")

plt.ylabel("Numbers")
c=['red','magenta','green','orange','blue','brown']
plt.scatter(x,y,color=c,s=200)
plt.show()
```

convert Image in gray colour 

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'hiii.jpg'

#opening emage using Pil

image = Image.open(fname).convert('L')

#mapping imamge to gray scale
plt.imshow(image,cmap='gray')
plt.show()
```

Pie chart

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['English','Hindi','Science','Math']
c=['Yellow','Brown','Blue','Black']
plt.pie(x,labels=y,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=['English','Hindi','Science','Math']

plt.pie(x,labels=y,colors=c)
plt.legend()
plt.show()
```
