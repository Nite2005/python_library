SEABORN

SEABORN is used for making statistical graph in python 
means used for visualisation .

it is interrelated on matplotlib and pandas


seaborn divides the plot into the below categories:

Relational plots : this plot is used to understand the realtion between two variables

categorical plots: this plot deals with categorical variables and how they can be visualized.

distribution plots : this plot is used for examining univariate and bivariate distributions;




```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
var = [1,2,3,4,5,6,7,8]
var1 =[8,7,6,5,4,3,2,1]
```

```python

plt.xlabel(var)
plt.ylabel(var1)
plt.title("graph")
plt.plot(var,var1)
plt.show()
```

```python
var = [1,2,3,4,5,6,7,8]
var1 =[8,7,6,5,4,3,2,1]
```

```python
sns.lineplot(x=var,y=var1)
plt.show()
```

```python
var = [1,2,3,4,5,6,7,8]
var1 =[8,7,6,5,4,3,2,1]

import pandas as pd

df = pd.DataFrame({"var":var,'var1':var1})
sns.lineplot(x='var',y='var1',data=df)
plt.show()
```

Order parameter

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data= df1,hue='sex',order=order1)
plt.show()
```

Hue_Order

```python
order1 = ['Dream','Torgersen','Biscoe']
hue_order1=['Female','Male']
sns.barplot(x='island',y='bill_length_mm',data= df1,hue='sex',order=order1,hue_order=hue_order1)
plt.show()
```

```python
order1 = ['Dream','Biscoe','Torgersen']

sns.barplot(x='island',y='bill_length_mm',data=df1 ,order=order1,color='green')
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',palette='Accent')
```

Saturation

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',palette='Accent',saturation=0.8)
```

Histogram is seaborn library

```python
df1=sns.load_dataset('penguins')
```

```python
df1
```

```python
sns.displot(df1['flipper_length_mm'])
plt.show()
```

for interval we use bins parameter

Bins parameter

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230])
```

color

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230],color='g')
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230],color='r')
```

histogram kernel density(kde)

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,190,200,210,220,230] , kde=True)
plt.show()
```
